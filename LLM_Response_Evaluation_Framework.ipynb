### LLM Response Evaluation Framework

This project will show you how to create a simple system to evaluate the quality of responses generated by a Large Language Model (LLM). We will use basic Python to define criteria and check if LLM responses meet them.

#### Step 1 – Project Goal

The goal of this project is to build a basic framework to automatically score LLM responses based on predefined rules. This helps in understanding how well an LLM performs on specific tasks without manual inspection of every response.

#### Step 2 – Install libraries if needed

For this simple project, we won't need to install any special libraries as we'll stick to built-in Python functionalities and `pandas` for data handling, which is usually pre-installed in Colab.

#### Step 3 – Import libraries

We will import the `pandas` library to easily manage our data in a structured format (like a table).

```python
import pandas as pd # Import the pandas library, which is great for working with data tables.
```

This code imports the `pandas` library and gives it a shorter name, `pd`, so we can use it more easily throughout our project.

#### Step 4 – Input data

We'll create a small dataset of sample prompts, LLM responses, and the expected keywords that a 'good' response should contain. This simulates real data you might get from an LLM.

```python
# Define our small dataset as a list of dictionaries
data = [
    {
        'prompt': 'What is the capital of France?',
        'llm_response': 'The capital of France is Paris. It is a beautiful city.',
        'expected_keywords': ['Paris', 'France', 'capital']
    },
    {
        'prompt': 'Explain photosynthesis briefly.',
        'llm_response': 'Photosynthesis is how plants make food using sunlight.',
        'expected_keywords': ['plants', 'food', 'sunlight']
    },
    {
        'prompt': 'Tell me about giraffes.',
        'llm_response': 'Giraffes are tall animals found in Africa.',
        'expected_keywords': ['tall', 'Africa']
    },
    {
        'prompt': 'What is the largest ocean?',
        'llm_response': 'The largest ocean is the Atlantic Ocean.', # This is incorrect for demonstration
        'expected_keywords': ['Pacific', 'largest', 'ocean'] # Expected for a correct answer
    }
]

# Convert the list of dictionaries into a pandas DataFrame for easier handling
df = pd.DataFrame(data) # Create a DataFrame from our data.

# Display the first few rows of our DataFrame to see what it looks like
print("Input Data:")
display(df) # Show the DataFrame content.
```

Here, we created a list of dictionaries, where each dictionary represents a scenario with a `prompt`, an `llm_response`, and a list of `expected_keywords`. We then converted this list into a `pandas` DataFrame, which is a tabular data structure that helps organize and display the data neatly. `display(df)` shows the contents of this table.

#### Step 5 – Main logic (small loop)

Now, we'll implement a simple evaluation logic. For each LLM response, we'll check how many of the `expected_keywords` are present in it. This will give us a simple score.

```python
results = [] # Create an empty list to store our evaluation results.

# Loop through each row (each prompt-response pair) in our DataFrame
for index, row in df.iterrows(): # `iterrows()` allows us to go through each row of the DataFrame.
    response = row['llm_response'] # Get the LLM's response for the current row.
    expected = row['expected_keywords'] # Get the list of expected keywords.
    prompt = row['prompt'] # Get the original prompt.

    score = 0 # Initialize the score for the current response to 0.
    found_keywords = [] # List to store keywords found in the response.

    # Loop through each keyword we expect to find
    for keyword in expected: # Iterate over each keyword in the 'expected_keywords' list.
        # Check if the keyword (case-insensitive) is in the LLM's response
        if keyword.lower() in response.lower(): # Convert both keyword and response to lowercase for a flexible match.
            score += 1 # If found, increase the score by 1.
            found_keywords.append(keyword) # Add the found keyword to our list.

    # Store the evaluation results for this response
    results.append({
        'prompt': prompt,
        'llm_response': response,
        'expected_keywords': expected,
        'found_keywords': found_keywords,
        'score': score,
        'max_score': len(expected) # The maximum possible score is the number of expected keywords.
    })

# Convert the results list into a new DataFrame for easy viewing
evaluation_df = pd.DataFrame(results) # Create a DataFrame from the collected results.
```

In this step, we iterate through each prompt and its LLM response in our `df`. For each response, we compare it against a list of `expected_keywords`. If a keyword is found in the response (ignoring case), we increment a `score`. Finally, we store all these details, including the prompt, response, expected keywords, found keywords, and the calculated score, into a new DataFrame called `evaluation_df`.

#### Step 6 – Output results

We will display the evaluation results and save them to a simple CSV (Comma Separated Values) file. This is a common format for storing tabular data.

```python
print("\nEvaluation Results:") # Print a heading for our results.
display(evaluation_df) # Display the DataFrame containing all evaluation results.

# Save the evaluation results to a CSV file
output_filename = 'llm_evaluation_results.csv' # Define the name of our output file.
evaluation_df.to_csv(output_filename, index=False) # Save the DataFrame to a CSV file. `index=False` prevents writing the DataFrame index as a column.

print(f"\nEvaluation results saved to {output_filename}") # Inform the user where the file is saved.
```

This code block first displays the `evaluation_df` so you can see all the calculated scores and details. Then, it saves this DataFrame into a file named `llm_evaluation_results.csv`. Saving to a CSV allows you to easily open and analyze the data later using spreadsheet software.

#### Step 7 – Short explanation of the result

From the `evaluation_df`, you can see how well each LLM response performed against the `expected_keywords`. For example, the first two responses scored perfectly (3/3), the third one scored 2/2, while the last one only scored 2/3 because it mentioned 'Atlantic Ocean' instead of 'Pacific', which was the expected keyword. This simple framework helps quickly identify strengths and weaknesses in an LLM's responses based on specific criteria.
